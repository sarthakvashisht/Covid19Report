```python
statewise = pd.pivot_table(df, values = ["Confirmed", "Deaths", "Cured"],index = "State/UnionTerritory", aggfunc = max)
statewise["Recovery_Rate"] = statewise["Cured"]*100 / statewise["Confirmed"]
statewise["Mortality_Rate"] = statewise["Deaths"]*100 / statewise["Confirmed"]
statewise['Active_Cases']=statewise['Confirmed']-(statewise['Cured']+statewise['Deaths'])
statewise = statewise.sort_values(by =  "Confirmed", ascending = False) 
statewise.style.background_gradient()
```

```python
df['Date']= pd.to_datetime(df['Date'], format = '%Y-%m-%d')
print(df)
```

```python
#cured
fig = plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df[df['State/UnionTerritory'].isin(['Delhi', 'Uttar Pradesh', 'Haryana', 'Uttarakhand', 'Rajasthan'])], x='Date', y='Cured', hue='State/UnionTerritory')
ax.set_title(" Rate of Cured cases in Delhi and Boundary Sharing States", size=20)
```

```python
#deaths
fig = plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df[df['State/UnionTerritory'].isin(['Delhi', 'Uttar Pradesh', 'Haryana', 'Uttarakhand', 'Rajasthan'])], x='Date', y='Deaths', hue='State/UnionTerritory')
ax.set_title(" Rate of Deaths cases in Delhi and Boundary Sharing States", size=20)
```

```python
#confirmed cases
fig = plt.figure(figsize=(15,8))
ax = sns.lineplot(data=df[df['State/UnionTerritory'].isin(['Delhi', 'Uttar Pradesh', 'Haryana', 'Uttarakhand', 'Rajasthan'])], x='Date', y='Confirmed', hue='State/UnionTerritory')
ax.set_title(" Rate of Confirmed cases in Delhi and Boundary Sharing States", size=20)
```
